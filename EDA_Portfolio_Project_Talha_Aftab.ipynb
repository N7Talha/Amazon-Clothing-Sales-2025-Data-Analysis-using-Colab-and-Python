# Initial Data exploration and Cleaning

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv("/content/Amazon_Clothing_Sales_2025_cleaned.csv")

print('Initial shape:')
print(df.shape)

```

```python
 #1. Schema & Types
print('Column dtypes before:')
print(df.dtypes)
df['order_date'] = pd.to_datetime(df['order_date'])
print('Converted order_date to datetime')
```

```python
# 2. Missing values
missing_counts = df.isna().sum().sort_values(ascending=False)
missing_pct = (missing_counts / len(df) * 100).round(2)
missing_df = pd.DataFrame({'missing_count': missing_counts, 'missing_pct': missing_pct})
print('Missing values per column:')
print(missing_df.head(10))
```

```python
# Strategy is to drop rows with critical fields missing (order_date, price, quantity, final_price)
critical_cols = ['order_date', 'price', 'quantity', 'final_price']
clean_df = df.dropna(subset=critical_cols).copy()
print('Shape after dropping rows with critical nulls:')
print(clean_df.shape)

# Identify non-critical columns
non_critical_cols = [col for col in clean_df.columns if col not in critical_cols]

# Impute missing values in non-critical columns
for col in non_critical_cols:
    if pd.api.types.is_numeric_dtype(clean_df[col]):
        # Fill numeric NaNs with mean
        clean_df[col].fillna(clean_df[col].mean(), inplace=True)
    else:
        # Fill categorical/text NaNs with mode
        mode_value = clean_df[col].mode()
        if not mode_value.empty:
            clean_df[col].fillna(mode_value[0], inplace=True)

print('Missing values after imputation:')
print(clean_df.isna().sum())
```

```python
# 3. Duplicates & Integrity
dup_count = clean_df.duplicated(subset=['order_id']).sum()
print('Duplicate order_id count:')
print(dup_count)
```

```python
# Validate final_price vs price, discount_percent, quantity
# Expected final price for the order: price * quantity * (1 - discount/100)
calc_final = clean_df['price'] * clean_df['quantity'] * (1 - clean_df['discount_percent'] / 100.0)
# tolerance for rounding/fees
final_diff = np.abs(clean_df['final_price'] - calc_final)
clean_df['price_calc_diff'] = final_diff
price_mismatch_pct = (final_diff > 0.05).mean() * 100
print('Percent of rows where final_price mismatches calculated (> 0.05):')
print(round(price_mismatch_pct,2))
```

```python
# 4. Outliers - using IQR for numeric columns
num_cols = ['price','final_price','discount_percent','quantity','delivery_days']
num_cols = [c for c in num_cols if c in clean_df.columns]
q1 = clean_df[num_cols].quantile(0.25)
q3 = clean_df[num_cols].quantile(0.75)
iqr = q3 - q1
lower_bounds = q1 - 1.5 * iqr
upper_bounds = q3 + 1.5 * iqr
outlier_flags = {}
for c in num_cols:
    outlier_flags[c] = ((clean_df[c] < lower_bounds[c]) | (clean_df[c] > upper_bounds[c])).sum()
print('Outlier counts by column:')
print(pd.Series(outlier_flags))
```

```python
# 5. Feature Engineering
clean_df['order_year'] = clean_df['order_date'].dt.year
clean_df['order_month'] = clean_df['order_date'].dt.to_period('M').astype(str)
clean_df['order_weekday'] = clean_df['order_date'].dt.day_name()
clean_df['discount_amount'] = clean_df['price'] * clean_df['discount_percent'] / 100.0
# Avoid division by zero
clean_df['unit_price'] = np.where(clean_df['quantity'] > 0, clean_df['final_price'] / clean_df['quantity'], np.nan)
# Delivery speed flag
clean_df['delivery_speed'] = np.where(clean_df['delivery_days'] <= 3, 'fast', 'slow')
# Customer-level aggregates
cust_grp = clean_df.groupby('customer_id').agg(
    total_spend = ('final_price','sum'),
    total_orders = ('order_id','nunique'),
    avg_rating_given = ('review_rating','mean'),
    returns = ('is_returned','sum')
).reset_index()
cust_grp['return_rate'] = np.where(cust_grp['total_orders']>0, cust_grp['returns'] / cust_grp['total_orders'], np.nan)
print('Engineered features added and customer aggregates computed')
```

# Section E: Advanced Analysis

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, roc_auc_score
import os
# Add order_month
df['order_month'] = df['order_date'].dt.to_period('m').astype(str)

# 2) Cohort analysis (fixed month arithmetic)

cohort_df = df[['customer_id','order_date']].dropna().copy()
cohort_df['order_date'] = pd.to_datetime(cohort_df['order_date'], errors='coerce')
cohort_df = cohort_df.dropna(subset=['order_date'])
cohort_df['order_year'] = cohort_df['order_date'].dt.year
cohort_df['order_month'] = cohort_df['order_date'].dt.month

firsts = cohort_df.groupby('customer_id')[['order_year','order_month']].min().reset_index().rename(columns={'order_year':'cohort_year','order_month':'cohort_month'})
cohort_df = cohort_df.merge(firsts, on='customer_id', how='left')
cohort_df['period_number'] = (cohort_df['order_year'] - cohort_df['cohort_year']) * 12 + (cohort_df['order_month'] - cohort_df['cohort_month'])

retention = cohort_df.groupby(['cohort_year','cohort_month','period_number'])['customer_id'].nunique().reset_index()
cohort_sizes = retention[retention['period_number'] == 0][['cohort_year','cohort_month','customer_id']].rename(columns={'customer_id':'cohort_size'})
retention = retention.merge(cohort_sizes, on=['cohort_year','cohort_month'], how='left')
retention['retention_rate'] = retention['customer_id'] / retention['cohort_size']

# Pivot to matrix
retention_matrix = retention.pivot_table(index=['cohort_year','cohort_month'], columns='period_number', values='retention_rate', fill_value=0.0)
retention_matrix.to_csv('cohort_retention_matrix.csv')
print('Saved cohort_retention_matrix.csv')

# Show a quick heatmap
plt.figure(figsize=(10,5))
sns.heatmap(retention_matrix, annot=False, cmap='Blues')
plt.title('Cohort Retention Rates')
plt.ylabel('Cohort (Y, M)')
plt.xlabel('Period Number (months)')
plt.tight_layout()
plt.show()
print('Displayed cohort retention heatmap')

# 3) RFM Segmentation + KMeans clustering
rfm_df = df.copy()
rfm_df['order_date'] = pd.to_datetime(rfm_df['order_date'], errors='coerce')
rfm_df = rfm_df.dropna(subset=['order_date'])
reference_date = rfm_df['order_date'].max() + pd.Timedelta(days=1)
agg = rfm_df.groupby('customer_id').agg({
    'order_date': lambda x: (reference_date - x.max()).days,
    'order_id': 'nunique',
    'final_price': 'sum'
}).reset_index()
agg = agg.rename(columns={'order_date':'recency','order_id':'frequency','final_price':'monetary'})

# Scale and cluster
rfm_features = agg[['recency','frequency','monetary']].replace([np.inf,-np.inf], np.nan).fillna(0.0)
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm_features)

kmeans = KMeans(n_clusters=4, random_state=42)
agg['rfm_cluster'] = kmeans.fit_predict(rfm_scaled)
agg.to_csv('customer_rfm_clusters.csv', index=False)
print('Saved customer_rfm_clusters.csv')

# 4) Simple price elasticity by category using log-log regression (price vs. quantity), per sub_category if available
elasticity_rows = []
if 'sub_category' in df.columns:
    cats = df['sub_category'].dropna().unique().tolist()
else:
    cats = ['ALL']
    df['sub_category'] = 'ALL'

for cat in cats:
    sub = df[df['sub_category'] == cat].copy()
    sub = sub[(sub['price'] > 0) & (sub['quantity'] > 0)]
    if len(sub) < 30:
        continue
    sub['log_q'] = np.log(sub['quantity'])
    sub['log_p'] = np.log(sub['price'])
    # Elasticity is slope of log_q on log_p
    x = sub['log_p'].values
    y = sub['log_q'].values
    if len(x) < 2:
        continue
    slope = np.polyfit(x, y, 1)[0]
    elasticity_rows.append({'sub_category': cat, 'price_elasticity': slope})

elasticity_df = pd.DataFrame(elasticity_rows)
elasticity_df.to_csv('price_elasticity_by_subcategory.csv', index=False)
print('Saved price_elasticity_by_subcategory.csv')

# 5) Return prediction (Logistic Regression and Decision Tree)
model_df = df.copy()

# Feature engineering minimal
model_df['discount_percent'] = pd.to_numeric(model_df['discount_percent'], errors='coerce')
model_df['delivery_days'] = pd.to_numeric(model_df['delivery_days'], errors='coerce')
model_df['review_rating'] = pd.to_numeric(model_df['review_rating'], errors='coerce')
model_df['final_price'] = pd.to_numeric(model_df['final_price'], errors='coerce')
model_df['quantity'] = pd.to_numeric(model_df['quantity'], errors='coerce')

# Encode categoricals simply with one-hot
cat_cols = []
for c in ['main_category','sub_category','brand','payment_method','region','customer_age_group','device_type']:
    if c in model_df.columns:
        cat_cols.append(c)

features = ['discount_percent','delivery_days','review_rating','final_price','quantity'] + cat_cols
model_df = model_df.dropna(subset=['is_returned'])
X_full = model_df[features].copy()
X_full = pd.get_dummies(X_full, columns=cat_cols, drop_first=True)
y = model_df['is_returned'].astype(int)

# Train/valid split
X_train, X_test, y_train, y_test = train_test_split(X_full, y, test_size=0.25, random_state=42, stratify=y)

# Logistic Regression
logreg = LogisticRegression(max_iter=200)
logreg.fit(X_train, y_train)
probs_lr = logreg.predict_proba(X_test)[:,1]
auc_lr = roc_auc_score(y_test, probs_lr)
report_lr = classification_report(y_test, logreg.predict(X_test), output_dict=True)

coef_series = pd.Series(logreg.coef_[0], index=X_full.columns).sort_values(key=np.abs, ascending=False).head(20)
coef_df = coef_series.reset_index()
coef_df.columns = ['feature','logistic_coefficient']
coef_df.to_csv('return_model_logreg_top_features.csv', index=False)
print('Saved return_model_logreg_top_features.csv')

# Decision Tree
tree = DecisionTreeClassifier(max_depth=5, random_state=42)
tree.fit(X_train, y_train)
probs_tr = tree.predict_proba(X_test)[:,1]
auc_tr = roc_auc_score(y_test, probs_tr)
report_tr = classification_report(y_test, tree.predict(X_test), output_dict=True)

imp_series = pd.Series(tree.feature_importances_, index=X_full.columns).sort_values(ascending=False).head(20)
imp_df = imp_series.reset_index()
imp_df.columns = ['feature','tree_importance']
imp_df.to_csv('return_model_tree_top_features.csv', index=False)
print('Saved return_model_tree_top_features.csv')

print('LogReg AUC: ' + str(round(auc_lr, 4)))
print('Tree AUC: ' + str(round(auc_tr, 4)))

# 6) Anomaly detection: flag orders with unusually high final_price or quantity by z-score
anom_df = df[['order_id','customer_id','final_price','quantity','price','discount_percent','review_rating','delivery_days']].copy()
for col in ['final_price','quantity']:
    vals = anom_df[col]
    z = (vals - vals.mean()) / (vals.std(ddof=0) if vals.std(ddof=0) != 0 else 1.0)
    anom_df[col + '_z'] = z

anomalies = anom_df[(anom_df['final_price_z'].abs() > 3) | (anom_df['quantity_z'].abs() > 3)].copy()
anomalies.to_csv('anomalous_orders.csv', index=False)
print('Saved anomalous_orders.csv')

# 7) Delivery hotspots by region and category: average delivery_days and return_rate
hot_df = df.copy()
if 'delivery_days' in hot_df.columns:
    hot_df['delivery_days'] = pd.to_numeric(hot_df['delivery_days'], errors='coerce')
if 'is_returned' in hot_df.columns:
    hot_df['is_returned'] = pd.to_numeric(hot_df['is_returned'], errors='coerce')

by_cols = []
for c in ['region','main_category']:
    if c in hot_df.columns:
        by_cols.append(c)
if len(by_cols) == 0:
    hot_df['region'] = 'ALL'
    hot_df['main_category'] = 'ALL'
    by_cols = ['region','main_category']

hot_agg = hot_df.groupby(by_cols).agg(
    avg_delivery_days = ('delivery_days','mean'),
    return_rate = ('is_returned','mean'),
    orders = ('order_id','nunique')
).reset_index()

hot_agg.to_csv('delivery_hotspots_by_region_category.csv', index=False)
print('Saved delivery_hotspots_by_region_category.csv')

# Quick barplot for top 10 slowest delivery combos
hot_plot = hot_agg.dropna(subset=['avg_delivery_days']).sort_values('avg_delivery_days', ascending=False).head(10)
plt.figure(figsize=(10,5))
sns.barplot(data=hot_plot, x='avg_delivery_days', y=by_cols[0], hue=by_cols[1])
plt.title('Top 10 Slowest Delivery Combos')
plt.xlabel('Average Delivery Days')
plt.ylabel(by_cols[0])
plt.tight_layout()
plt.show()
print('Displayed hotspots barplot')

```
